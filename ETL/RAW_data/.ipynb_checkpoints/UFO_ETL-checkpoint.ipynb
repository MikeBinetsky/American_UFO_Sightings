```python
## Import dependencies for API call
import pandas as pd
import csv
```

```python
## Importing filepath for raw data
filePath = "ufo_sightings_scrubbed.csv"
```

```python
## Importing our CSV into a pandas dataframe
ufoSightnings = pd.read_csv(filePath)
```

```python
## view the dataframe
ufoSightnings.head()
```

```python
## Sort dataframe to only us sightings
usOnlySightings = ufoSightnings[ufoSightnings.country == "us"]
```

```python
usOnlySightings
```

```python
# split datetime column to get a year column
usOnlySightings['year'] = usOnlySightings['datetime'].str.split('/').str[2]
usOnlySightings['year'] = usOnlySightings['year'].str.split(' ').str[0]
```

```python
usOnlySightings
```

```python
## Convert time column into datetime for easier parsing
usOnlySightings["datetime"] = pd.to_datetime(usOnlySightings["datetime"])
```

```python
## Confirming
usOnlySightings.dtypes
```

```python
usOnlySightings
```

```python
usOnlySightings.to_csv("../Cleaned_data/ufo_sightings_cleaned.csv")
```
