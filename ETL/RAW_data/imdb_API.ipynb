```python
## Importing dependencies
import pandas as pd
import requests
import json
import csv

## import API key
from API_keys import imdb_api
```

```python
## Setting up API call
apiKey = imdb_api
## Getting base URL
base_url = f'https://imdb-api.com/en/API/Keyword/{apiKey}/'

## Setting up a list of relevant search terms for API call
apiSearch = ["alien", "UFO", "Alien Invasion", "Extraterrestrial", "Unidentified Flying Object", "Alien Abduction", "Alien Creature", "Flying Saucer", "Alien Technology", "Alien Race"]
```

```python
## Performing an API call loop.

## This empty list will store all the results
apiResults = []

## Loop through each entry in the api search list
for search in apiSearch:
    apiURL = base_url + search
    response = requests.get(apiURL)
    responseJson = response.json()
    responseList = responseJson["items"]
    for movie in responseList:
        ## This creates a list of all the items in the list of dictionaries that is apiResults
        #list_of_all_values = [value for elem in apiResults for value in elem.values()]
        
        if movie["id"] not in apiResults:
            apiResults.append(movie)
        else:
            pass
```
